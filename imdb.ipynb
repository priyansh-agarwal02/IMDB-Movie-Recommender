```python
mov=mov[['Poster_Link', 'Series_Title', 'Genre', 'Overview']]
```

```python
mov
```

```python
mov['tags'] = mov['Overview']+mov['Genre']
mov
```

```python
new_data  = mov.drop(columns=['Overview', 'Genre'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
new_data[new_data['Series_Title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].Series_Title)
```

```python
def recommand(movies):
    index=new_data[new_data['Series_Title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].Series_Title)
```

```python
recommand("Iron Man")
```
